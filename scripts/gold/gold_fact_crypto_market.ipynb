##creating table

```python
query="""
SELECT
    id as coin_id,
    ingestion_date,
    current_price AS price,
    market_cap,
    total_volume AS volume,
    round(price_vs_ath,4) AS price_vs_ath,
    round(volume_to_market_cap,4) AS volume_to_market_cap,
    round(circulating_supply_ratio,4) AS circulating_supply_ratio

FROM workspace.silver.crypto
"""

df=spark.sql(query)

print("created successfully!")
```

##checking table

```python
display(df.limit(5))
```

##saving

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_crypto_market")
print("saved successfully!")
```
