##creating table

```python
query="""
SELECT DISTINCT
    ingestion_date AS date_id,
    day(ingestion_date) AS day,
    month(ingestion_date) AS month,
    year(ingestion_date) AS year,
    quarter(ingestion_date) AS quarter,
    dayofweek(ingestion_date) AS day_of_week,
    weekofyear(ingestion_date) AS week_of_year
FROM workspace.silver.crypto
"""

df=spark.sql(query)

print("created successfully!")
```

##checking table

```python
display(df.limit(5))
```

##saving table

```python
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").saveAsTable("workspace.gold.dim_date")

print("saved successfully!")
```
