##creating table

```python
query="""
SELECT DISTINCT
    id,
    name,
    symbol,
    market_cap_rank
FROM workspace.silver.crypto
order by market_cap_rank asc
"""

df=spark.sql(query)

print("created successfully!")
```

##checking table

```python
display(df.limit(5))
```

##saving table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_coin")

print("saved successfully!")
```
