##Use Catalog workspace

```python
%sql
USE CATALOG workspace;
```

##Create Schemas

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'bronze layer (raw data)';

CREATE SCHEMA IF NOT EXISTS silver
COMMENT 'silver layer (cleaned/transformed)';

CREATE SCHEMA IF NOT EXISTS gold
COMMENT 'gold layer (business ready)';
```

##Create Volume

```python
%sql
CREATE VOLUME IF NOT EXISTS workspace.bronze.raw_source
COMMENT 'volume for raw data as json format';
```
